```python
from datetime import datetime
reference_year = 2017
DF3['Age'] = reference_year - DF3['birthyear']
```

```python
DF3['start_time'] = pd.to_datetime(DF3['start_time'])
DF3['end_time'] = pd.to_datetime(DF3['end_time'])
```

```python
DF3['Month'] = DF3['start_time'].dt.month
```

```python
DF3['Start Date'] = DF3['start_time'].dt.date
DF3['End Date'] = DF3['end_time'].dt.date
DF3['Date'] = DF3['end_time'].dt.date
DF3['Day'] = DF3['start_time'].dt.day
DF3['Start Hour'] = DF3['start_time'].dt.hour
DF3['End Hour'] = DF3['end_time'].dt.hour
```
