```python
# University Ranking
QS = pd.read_excel('/content/drive/MyDrive/Modelling analytics/Modelling Analytics Project/QS_Asia.xlsx')

import numpy as np

for index, row in df.iterrows():
    matching_rows = QS[QS['Name'] == row['University_name']]

    if not matching_rows.empty:
        rank = matching_rows.iloc[0]['Ranking']


        if 1 <= rank <= 100:
            df.at[index, 'University'] = 9
        elif 101 <= rank <= 200:
            df.at[index, 'University'] = 8
        elif 201 <= rank <= 300:
            df.at[index, 'University'] = 7
        elif 301 <= rank <= 400:
            df.at[index, 'University'] = 6
        elif 401 <= rank <= 500:
            df.at[index, 'University'] = 5
        elif 501 <= rank <= 600:
            df.at[index, 'University'] = 4
        elif 601 <= rank <= 700:
            df.at[index, 'University'] = 3
        elif 701 <= rank <= 800:
            df.at[index, 'University'] = 2
        elif 801 <= rank <= 900:
            df.at[index, 'University'] = 1
    else:
        df.at[index, 'University'] = 1
```

```python
#Academic - Major
df['Major'] = df['Major'].str.lower()
substring1 = 'operational'
substring2 = 'science'
substring3 = 'math'
substring4 = 'statis'
substring5 = 'financ'
substring6 = 'econo'
substring7 = 'physics'
substring8 = 'engineer'
substring9 = 'bio'
substring10 = 'tech'
substring11 = 'info'
substring12 = 'life'
substring13 = 'data'
substring14 = 'analy'
substring15 = 'business'
substring16 = 'computer'
substring17 = 'accounting'
df['Major'] = df['Major'].apply(lambda x: 1 if (substring1 in x or substring2 in x or substring3 in x
                                                    or substring4 in x or substring5 in x or substring6 in x
                                                    or substring7 in x or substring8 in x or substring9 in x
                                                    or substring10 in x or substring11 in x or substring12 in x
                                                    or substring13 in x or substring15 in x or substring16 in x or substring17 in x) else x)
df['Major'] = df['Major'].apply(lambda x: 0 if x != 1 else x)
```

```python
# #Academic - GPA
mapping = {'Above 85%': 5, '80%~85%': 4, '75%~80%': 3,'70%~75%':2,'Below 70%':1}

df['GPA'] = df['Please enter your GPA of the major'].map(mapping)
```

```python
#Academic - Additional degree
df['Addition_degree'] = df['Do you have more than one degree instead of only bachelor? If yes, please choose the degree you have.'].map({
    "No, I don't have other degrees": 0,
    'Master': 1,
    'PhD': 2})
```

##Work Experience

```python
#Work Experience (Start time & End time) - Duration & End year
#Work time
df['Start time'] = df['Start time'].fillna("10/10/1900")
df['End time'] = df['End time'].fillna("10/10/1900")

df['Start time'] = pd.to_datetime(df['Start time'])
df['End time'] = pd.to_datetime(df['End time'])

df['Work_duration'] = (df['End time'] - df['Start time']).dt.days / 365.25
#df_1['Work_duration'] = df_1['Work_duration'].apply(lambda x: 1 if x <= 1 else (2 if x <= 3 else (3 if x <= 6 else 4)))
df['Work_duration'] = df['Work_duration'].apply(lambda x: 0 if x == 0 else (1 if x <= 1 else (2 if x <= 3 else (3 if x <= 6 else 4))))

df['End_year'] = df['End time'].dt.year
df['End_year'] = df['End_year'].apply(lambda x: 4 if x == 2023 else (3 if x == 2022 else (2 if x == 2021 else (0 if x== 1990 else 1))))
```

```python
for index, row in df.iterrows():
  if row['Work_duration'] in [3,4]:
    if row['End_year'] == 4:
      df.at[index, 'Work_time'] = 8
    elif row['End_year'] == 3:
      df.at[index, 'Work_time'] = 7
    elif row['End_year'] == 2:
      df.at[index, 'Work_time'] = 6
    elif row['End_year'] == 1:
      df.at[index, 'Work_time'] = 5
  elif row['Work_duration'] in [1,2]:
    if row['End_year'] == 4:
      df.at[index, 'Work_time'] = 4
    elif row['End_year'] == 3:
      df.at[index, 'Work_time'] = 3
    elif row['End_year'] == 2:
      df.at[index, 'Work_time'] = 2
    elif row['End_year'] == 1:
      df.at[index, 'Work_time'] = 1
  else:
    df.at[index, 'Work_time'] = 0
```

```python
#Company name
Forbes = pd.read_excel('/content/drive/MyDrive/Modelling analytics/Modelling Analytics Project/Forbes 2000 List.xlsx')

for index, row in df.iterrows():
    matching_rows = Forbes[Forbes['Company_name'] == row['Company_name']]

    if not matching_rows.empty:
        rank = matching_rows.iloc[0]['Rank']

        if rank <= 200:
            df.at[index, 'Company_name'] = 9
        elif 201 <= rank <= 400:
            df.at[index, 'Company_name'] = 8
        elif 401 <= rank <= 600:
            df.at[index, 'Company_name'] = 7
        elif 601 <= rank <= 800:
            df.at[index, 'Company_name'] = 6
        elif 801 <= rank <= 1000:
            df.at[index, 'Company_name'] = 5
        elif 1001 <= rank <= 1200:
            df.at[index, 'Company_name'] = 4
        elif 1201 <= rank <= 1400:
            df.at[index, 'Company_name'] = 3
        elif 1401 <= rank <= 1600:
            df.at[index, 'Company_name'] = 2
        elif 1601 <= rank <= 1800:
            df.at[index, 'Company_name'] = 1
    else:
        df.at[index, 'Company_name'] = 1

df.head(5)
```

```python
#Work experience - Work position
df['Position1'] = df['Position1'].str.lower()
df['Work_position'] = df['Position1'].fillna("0")
substring1 = 'associate'
substring2 = 'manager'
substring3 = 'leadership'
substring4 = 'junior'
substring5 = 'intern'
df['Work_position'] = df['Work_position'].apply(lambda x: "4" if (substring2 in x) else x)
df['Work_position'] = df['Work_position'].apply(lambda x: "5" if (substring3 in x) else x)
df['Work_position'] = df['Work_position'].apply(lambda x: "3" if (substring1 in x) else x)
df['Work_position'] = df['Work_position'].apply(lambda x: "2" if (substring4 in x) else x)
df['Work_position'] = df['Work_position'].apply(lambda x: 5 if x == "5" else (4 if x == "4"
                                                                              else (3 if x == "3" else (2 if x=="2" else (0 if x=="0" else 1)))))
```

```python
# Work department
df['Work_department'] = df['Work_department'].str.lower().fillna("0")

substrings = {
    'market': 2,
    'scien': 6,
    'engineer': 5,
    'bank': 4,
    'finance': 3,
    'analy': 7,
    'statis': 7
}

def categorize_department(x):
    for substring, category in substrings.items():
        if substring in x:
            return category
    return 0

df['Work_department'] = df['Work_department'].apply(categorize_department)
```

```python
# Rescaling all the variables to 0-1
columns_to_scale_work = ['Company_name','Work_position','Work_department','Work_time']
data_to_scale_work = df[columns_to_scale_work]
scaler = MinMaxScaler()
scaled_data_work = scaler.fit_transform(data_to_scale_work)
num_decimals = 5
scaled_data_work_rounded = scaled_data_work.round(num_decimals)
df[columns_to_scale_work] = scaled_data_work_rounded
```
